```python
import xarray as xr
import pint
import pint_xarray
```

```python
name = "pacific_seapodym"
```

## Biomass

```python
data = xr.open_mfdataset(f"./{name}/output/*biomass*.nc").load()
```

```python
data["biomass"].dropna("time", how="all").mean(["latitude", "longitude"]).plot()
```

```python
data["biomass"].dropna("time", how="all").mean(["latitude", "longitude"])
```

```python
data = data.resample(time="1D").mean()
```

```python
data = data.rename({"latitude": "y", "longitude": "x"})
data.x.attrs = {}
data.y.attrs = {}
```

```python
data = data.pint.quantify().pint.to("g/m**2").pint.dequantify()
```

```python
data["biomass"].mean("time").plot()
```

```python
data.to_zarr(f"./{name}/result.zarr", mode="w")
```

## Recruitment

```python
recruitment = xr.open_mfdataset(f"./{name}/output/*recprod*.nc").load()
recruitment = recruitment["recprod"]
recruitment.name = "recruitment"
recruitment
```

```python
recruitment.dropna("time", how="all").mean(["latitude", "longitude"]).plot()
```

```python
recruitment = recruitment.resample(time="1D").mean()
```

```python
recruitment = recruitment.rename({"latitude": "y", "longitude": "x"})
recruitment.x.attrs = {}
recruitment.y.attrs = {}
```

```python
recruitment.attrs["units"] = "milligram / meter ** 2 / day"
```

```python
recruitment.mean("time").plot()
```

```python
recruitment
```

```python
recruitment.to_zarr(f"./{name}/recruitment.zarr", mode="w")
```
